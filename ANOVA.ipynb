```python
 import scipy.stats as stats
tech_a = [85,88,84,92,90]
tech_b = [78,82,80,85,88]
tech_c = [92,95,96,89,91]
f_stat,p_val = stats.f_oneway(tech_a,tech_b,tech_c)
alpha = 0.05
print('f_stat',f_stat)
print('p_val',p_val)
```

```python
if p_val < alpha:
    print('Reject null hypothesis:\nThere is significant differnce between mean scores of three techniques')
else:
    print('Do not reject null hypothesis:\nThere is no significant differnce between mean scores of three techniques')
```

```python
# TWO WAY ANOVA
import numpy as np 
import pandas as pd

df = pd.DataFrame({'water':np.repeat(['daily','weekly'],15),
                   'sun':np.tile(np.repeat(['low','med','high'],5),2),
                   'height':[6,6,6,5,6,5,5,6,4,5,
                             6,6,7,8,7,3,4,4,4,5,
                             4,4,4,4,4,5,6,6,7,8]})
df

```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols('height ~ C(water) + C(sun) +C(water):C(sun)',data=df).fit()
sm.stats.anova_lm(model,typ = 2)
```

```python
data = pd.DataFrame({'AgeGroup':['Age Grp 1']*15+['Age Grp 2']*15,
                     'Technique':['Tech A']*5+['Tech B']*5+['Tech C']*5+
                                ['Tech A']*5+['Tech B']*5+['Tech C']*5,
                     'Score':[85,88,84,90,92,78,82,80,85,88,92,95,89,96,91,
                              78,80,82,85,89,92,95,91,96,90,84,88,85,90,82]})
data
```

```python
formula = 'Score ~ C(AgeGroup)+C(Technique)+C(AgeGroup):C(Technique)'
model = ols(formula,data=data).fit()
anova_table = sm.stats.anova_lm(model,typ =2)
print(anova_table)
```
