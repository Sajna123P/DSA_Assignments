*5. VISUALIZATIONS*

```python
import seaborn as sns
plt.bar(iris['Classification'],iris['PL'],color = 'r',width = 0.2)
plt.title('PETAL LENGTH CLASSIFICATION')
plt.ylabel('PL')
print('Below is how petal length varies among the given three species of iris flower')


```

```python
import seaborn as sns
plt.figure(figsize=(10,6))
sns.countplot(x=iris.SL)
plt.xticks(rotation=90)
print('Below is shown where most values and least values of sepal length falls among the given three species of iris flower')

```

```python
print('Below is the value ranges of petal length,petal width,sepal length and sepal width of iris setosa species')
iris[['PL','PW','SL','SW']].head(50).plot()

```

```python
plt.scatter(iris['SL'],iris['SW'])
plt.title('Scatterplot of Sepal length vs Sepal width')
plt.xlabel('SL')
plt.ylabel('SW')
```

```python

sns.heatmap(data = iris.corr(),annot = True, cmap = 'RdYlGn')
```

From the heatmap, we come to know that the pairs that have a high correlation are:

Petal Length and Sepal Length

Sepal Length and Petal Width

Petal Length and Petal Width
