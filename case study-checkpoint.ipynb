### checking for null values..  

```python
data.isnull().sum()
```

### user_id and device_type are not so relevant, so dropping those columns

```python
data1 = data.drop('user_id',axis = 1,inplace= True)
data2 = data.drop('device_type',axis = 1,inplace= True)
```

### Filling columns that have null values,filling numerical columns with mode and non numerical column with ffill

```python
for i in ['session_id','item_id','item_price',
         'category','product_type','impression_time','app_code']:
     data[i] = data[i].fillna(data[i].median())
data['os_version'].fillna(method='ffill', inplace= True)
data.isnull().sum()
```

### Finding outliers and removing them

```python
import matplotlib.pyplot as plt
num_cols = ['item_price','category','product_type','app_code']
for i in num_cols:
    plt.figure()
    plt.boxplot(data[i])
    plt.title(i)
```

### It is found that item price has a lot of outliers, so removing them

```python
Q1 = np.percentile(data['item_price'],25,method = 'midpoint')
Q2 = np.percentile(data['item_price'],50,method = 'midpoint')
Q3 = np.percentile(data['item_price'],75,method = 'midpoint')
print('Q1 : ',q1,'\nQ2 : ',q2,'\nQ3 : ',q3)
```

```python
IQR = Q3-Q1
low_lim = Q1 -1.5*IQR
up_lim = Q3 +1.5*IQR
outlier =[]
for x in data ["item_price"]:
    if((x>up_lim)or(x<low_lim)):
        outlier.append(x)
outlier
```

```python
to_be_removed = data["item_price"]>up_lim
data.loc[to_be_removed].index
```

```python
data.drop(data.loc[to_be_removed].index,inplace = True)
```

### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder  #LABEL ENCODING

le = LabelEncoder()
from sklearn import preprocessing
std = preprocessing.StandardScaler()

data['os_version'] = le.fit_transform(data['os_version'])
data['server_time'] = le.fit_transform(data['server_time'])
data['impression_time'] = le.fit_transform(data['impression_time'])

# x is an array,convert it into DF
x = std.fit_transform(data)
x = pd.DataFrame(x)
x.describe()
```
