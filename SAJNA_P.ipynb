### 4. Fill in all the missing values present in all the columns in the dataset

```python
data['Cabin'].fillna(method='bfill', inplace= True)
data['Cabin'].fillna(method='ffill', inplace= True)
data['Embarked'].fillna(method='bfill', inplace= True)
data['Age'] = data['Age'].fillna(data['Age'].median())
data.head(50)
```

```python
print(data.isna().sum())
sns.heatmap(data.isnull(),yticklabels=False,cbar=False)
```

### 5. Check and handle outliers in at least 3 columns in the dataset
 

```python
num_cols = ['Age', 'SibSp', 'Parch','Fare']
for i in num_cols:
    plt.figure()
    plt.boxplot(data[i]) # boxplot to detect outliers
    plt.title(i)
```

### a) 'Sibsp',Parch ' contain natural number outliers below 10, so there is nothing unusual in it,so those outliers cannot be removed
### b)'Age' and 'Fare' contains a lot of outliers, so moving on to find the number and nature of outliers
### 'Age'

```python
Q1 = np.percentile(data['Age'],25,method = 'midpoint')
Q2 = np.percentile(data['Age'],50,method = 'midpoint')
Q3 = np.percentile(data['Age'],75,method = 'midpoint')
IQR = Q3-Q1
low_lim = Q1 -1.5*IQR
up_lim = Q3 +1.5*IQR
outlier =[]
for x in data ["Age"]:
    if((x>up_lim)or(x<low_lim)):
        outlier.append(x)
print('Lower limit: ',low_lim,'Upper limit: ',up_lim)
outlier
```

### There are some decimal values below the lower limit in the ouliers, those are set to lower limit

```python
data['Age'] = np.where(data['Age'] < low_lim, low_lim, data['Age'])
outlier =[]
for x in data ["Age"]:
    if((x>up_lim)or(x<low_lim)):
        outlier.append(x)
outlier
```

### Since all these are valid ages,they cannot  be neglected
### 'Fare'

```python
q1 = np.percentile(data['Fare'],25,method = 'midpoint')
q2 = np.percentile(data['Fare'],50,method = 'midpoint')
q3 = np.percentile(data['Fare'],75,method = 'midpoint')
IQR = q3-q1
low_lim = q1 -1.5*IQR
up_lim = q3 +1.5*IQR
outlier2 =[]
for x in data ["Fare"]:
    if((x>up_lim)or(x<low_lim)):
        outlier2.append(x)
outlier2
```

### As the cruise can have any fare ranges that falls in the outliers according to the classes and luxuries of the passengers, outliers cannot be removed.
